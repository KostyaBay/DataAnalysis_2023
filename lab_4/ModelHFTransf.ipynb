```python
import tensorflow as tf
```

```python
from transformers import pipeline
```

Calling the model for automatic recognition and generation of a text summary (summarization) of the text in the Ukrainian language

```python
summarizer = pipeline("summarization", model='ukr-models/uk-summarizer')
```

```python
text = "Записки українського самашедшого - це книга, яка переносить читача в унікальний світ автентичного відображення та глибокого розгортання української культури та психології.\n Книга розпочинається зі вступу, глибоко аналізуючи власні рефлексії та спостереження автора щодо навколишнього світу. Через кожен рядок читач відчуває нестримну жагу автора до виявлення сенсу існування та власної ідентичності."
```

```python
summary = summarizer(text, max_length=110, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)

print(summary[0]['summary_text'])
```

The function allows you to insert masked (hidden) tokens into the specified text and get probable variants for these tokens:

```python
unmasker = pipeline('fill-mask', model='xlm-roberta-base')
```

```python
text = "Роман - це літературний <mask>"
```

```python
unmasker(text)
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.utils import shuffle
```

### N_colors = 64

```python
n_colors = 64

scoob = plt.imread("Scoob_Poster.jpg") #load the Scooby-doo poster

# Convert to floats instead of the default 8 bits integer coding. Dividing by 255 is important so that plt. 
# It show behaves works well on float data (need to be in the range [0-1])
scoob = np.array(scoob, dtype=np.float64) / 255

# Load Image and transform to a 2D numpy array.
w, h, d = original_shape = tuple(scoob.shape)
assert d == 3
image_array = np.reshape(scoob, (w * h, d))

print("Fitting model on a small sub-sample of the data")
image_array_sample = shuffle(image_array, random_state=0, n_samples=1_000)
kmeans = KMeans(n_clusters=n_colors, n_init="auto", random_state=0).fit(
    image_array_sample
)
```

##### Get labels for all points

```python
print("Predicting color indices on the full image (k-means)")
labels = kmeans.predict(image_array)


codebook_random = shuffle(image_array, random_state=0, n_samples=n_colors)
print("Predicting color indices on the full image (random)")
labels_random = pairwise_distances_argmin(codebook_random, image_array, axis=0)


def recreate_image(codebook, labels, w, h):
    """Recreate the (compressed) image from the code book & labels"""
    return codebook[labels].reshape(w, h, -1)
```

##### How many different colors in the Scooby-doo poster?

```python
from PIL import Image

image_path = "Scoob_Poster.jpg" #load the Scoob poster
img = Image.open(image_path)

unique_colors = set(img.getdata()) #get the value of unique colors
length = len(unique_colors)
print(length) #show value of colors
```

##### So show all result, alongside original image

```python
plt.figure(1)
plt.clf()
plt.axis("off")
plt.title(f"Original image ({length} colors)")
plt.imshow(scoob)

plt.figure(2)
plt.clf()
plt.axis("off")
plt.title(f"Quantized image ({n_colors} colors, K-Means)")
plt.imshow(recreate_image(kmeans.cluster_centers_, labels, w, h))

plt.figure(3)
plt.clf()
plt.axis("off")
plt.title(f"Quantized image ({n_colors} colors, Random)")
plt.imshow(recreate_image(codebook_random, labels_random, w, h))
plt.show()
```
